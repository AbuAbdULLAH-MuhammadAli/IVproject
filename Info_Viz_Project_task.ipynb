```python
import csv


Isloo_List=[]
with open('outlierremovedZameen.csv') as file:
    reader=csv.reader(file)
    count=0
    cat1_total=0
    cat2_total=0
    cat3_total=0
    cat4_total=0
    cat5_total=0
    Isloo_0_5Marla_array=[]
    Isloo_5_10Marla_array=[]
    Isloo_10_15Marla_array=[]
    Isloo_15_20Marla_array=[]
    Isloo_1_5Kanal_array=[]
    
    for row in reader:
        if(row[6]=="Islamabad"):
            print(row[6]," " , row[4]," " ,row[11], row[19])
        
            if(row[19]=="0-5 Marla"):
                print("Category 1")
                Isloo_0_5Marla_array.append(row[4])


            elif(row[19]=="5-10 Marla"):
                print("Category 2")
                Isloo_5_10Marla_array.append(row[4])

            elif(row[19]=="10-15 Marla"):
                print("Category 3")
                Isloo_10_15Marla_array.append(row[4])

            elif(row[19]=="15-20 Marla"):
                print("Category 4")
                Isloo_15_20Marla_array.append(row[4])

            elif(row[19]=="1-5 Kanal"):
                print("Category 5")
                Isloo_1_5Kanal_array.append(row[4])
                   

#Calculating length of arrays            
length_cat1=len(Isloo_0_5Marla_array)
length_cat2=len(Isloo_5_10Marla_array)    
length_cat3=len(Isloo_10_15Marla_array)
length_cat4=len(Isloo_15_20Marla_array)
length_cat5=len(Isloo_1_5Kanal_array)

#Calculating the sums

for i in range(len(Isloo_0_5Marla_array)):
            b1=int(Isloo_0_5Marla_array[i])    
            cat1_total=cat1_total+b1

for i in range(len(Isloo_5_10Marla_array)):
            b2=int(Isloo_5_10Marla_array[i])    
            cat2_total=cat2_total+b2

for i in range(len(Isloo_10_15Marla_array)):
            b3=int(Isloo_10_15Marla_array[i])    
            cat3_total=cat3_total+b3

for i in range(len(Isloo_15_20Marla_array)):
            b4=int(Isloo_15_20Marla_array[i])    
            cat4_total=cat4_total+b4

for i in range(len(Isloo_1_5Kanal_array)):
            b5=int(Isloo_1_5Kanal_array[i])    
            cat5_total=cat5_total+b5

def calculate_average(addition,total):
    c=addition/total
    c=round(c)
    return c 

if(length_cat1>0):
    res1=calculate_average(cat1_total,length_cat1)
    print("Average Per Marla price of 0-5 Marla in Islamabad= "+str(res1))
    Isloo_List.append(res1)
if(length_cat2>0):
    res2=calculate_average(cat2_total,length_cat2)
    print("Average Per Marla price of 5-10 Marla in Islamabad= "+str(res2))
    Isloo_List.append(res2)
if(length_cat3>0):
    res3=calculate_average(cat3_total,length_cat3)
    print("Average Per Marla price of 10-15 Marla in Islamabad= "+str(res3))
    Isloo_List.append(res3)
if(length_cat4>0):
    res4=calculate_average(cat4_total,length_cat4)
    print("Average Per Marla price of 15-20 Marla in Islamabad= "+str(res4))
    Isloo_List.append(res4)
if(length_cat5>0):
    res5=calculate_average(cat5_total,length_cat5)
    print("Average Per Marla price of 1-5 Kanal in Islamabad= "+str(res5)) 
    Isloo_List.append(res5)
```
